## Movie Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

movies = pd.read_csv('movies.csv', encoding = "ISO-8859-1")
movies.head()
```

# 1) What was the run-time for each movie?

1) What was the run-time for each movie?

```python
plt.figure(figsize=[14,6])
plt.xticks(rotation=90)
plt.bar(movies['Movie Title'], movies['Runtime'], width=0.7)

plt.title('Movie Runtimes', fontsize=20)
plt.xlabel('Movie Title', fontsize=15)
plt.ylabel('Runtime(Minutes)', fontsize=15)

plt.show()
```

# 2) What was the movie with the longest run time?


As seen in the graph above, the movie with the longest runtime was '12 Years a Slave'(in the original dataset). In the Exceeds Expectations dataset, 'Drive My Car' is the movie with the longest runtime. I found this result by creating a bar chart using the matplotlib bar function.


# 3) Is there any relationship between the length of the movie (runtime) and the number of awards a movie won?


There appears to be no relationship between runtime and the number of awards a movie won as seen in the scatterplot below. To get this result I used matplotlib to create a scatterlot of Runtime versus Award wins.

```python
plt.figure(figsize=[14,7])
plt.scatter(movies['Runtime'], movies['Award Wins'])
plt.title('Relationship between Runtime and Award Wins')
plt.xlabel('Movie Runtime')
plt.ylabel('Award Wins')
plt.show()
```

# 4)How many awards and nominations did each movie win?


I answered this question by creating a graph with multiple bars per movie using the xticks() function.

```python
plt.figure(figsize=[14,7])
plt.xticks(rotation=90)

X_axis = np.arange(len(movies['Movie Title']))
plt.bar(X_axis + 0.2, movies['Award Wins'],0.4,label="Award Wins")
plt.bar(X_axis - 0.2, movies['Award Nominations'],0.4,label="Award Nominations")
plt.xticks(X_axis, movies['Movie Title'])
plt.legend()
plt.ylabel('Total', fontsize=16)
plt.xlabel('Movie Title', fontsize=16)
plt.title('Number of Award Wins and Award Nominations Each Movie Won', fontsize=19)
```

## Exceeds Expectations Questions


# 1) Is there a relationship between the amount of box office earnings a movie had and the amount of total nominations the movie received (total nominations = awards wins + award nominations)?


There appears to be no relationship between a movie's box office earnings and the total number of nominations that the movie received. I found this result by creating a scatterplot using the matplotlib scatterplot function.

```python
movies['Total Nominations'] = movies['Award Wins'] + movies['Award Nominations']
plt.figure(figsize=[14,7])
plt.scatter(movies['Total Nominations'], movies['Box Office'])
plt.title('Relationship between Total Nominations and Box Office Earnings', fontsize = 16)
plt.xlabel('Total Nominations', fontsize=16)
plt.ylabel('Box Office', fontsize=16)
plt.show()
```

# 2) Is there a relationship between box office earnings and movie runtimes?


There appears to be no relationship between a movie's box office earnings and the movie's runtime. I found this result by creating a scatterplot using the matplotlib scatterplot function.

```python
plt.figure(figsize=[14,7])
plt.scatter(movies['Runtime'],movies['Box Office'])
plt.title('Relationship between Box Office Earnings and Movie Runtimes')
plt.xlabel('Movie Runtime')
plt.ylabel('Box Office Earnings')
plt.show()
```

# 3) How much did each movie earn at the box office?


I created a bar chart below to illustrate how much money each movie earned at the box office. I used the matplotlib bar function to create this graph.

```python
plt.figure(figsize=[14,7])
plt.xticks(rotation=90)
plt.bar(movies['Movie Title'], movies['Box Office'])
plt.title('Amount of Money Each Movie Earned at the Box Office')
plt.xlabel('Movie Title')
plt.ylabel('Box Office Earnings')
```

# 4) What is the total count of each genre present in the dataset? (How many times does fantasy, drama, adventure, etc. show up)

The graph below shows the total count of each genre present in the dataset. In order to find this result, I create a loop that split each cell with the genres each movie was considered to fall under. Then I added each individul word or genre to a list. After the loop stopped, I used seaborn to create a bar chart that counted the number of times each word appeared in the list.

```python
genres = []
for row in movies.itertuples(): 
    for word in row.Genre.split(', '):
        genres.append(word)
plt.xticks(rotation=90)
ax=sns.countplot(genres)
plt.xlabel('Genre')
plt.title('Total Count of Each Genre')
plt.show()
```
